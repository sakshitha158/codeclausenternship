```python
import tkinter as tk
from tkinter import messagebox
import speech_recognition as sr
import pyttsx3

class VoiceAssistant:
    def __init__(self, root):
        self.root = root
        self.root.title("Voice Assistant")
        self.root.geometry("400x300")

        self.label = tk.Label(self.root, text="Press the button and speak", font=("Helvetica", 14))
        self.label.pack(pady=20)

        self.text_box = tk.Text(self.root, width=50, height=10, wrap='word')
        self.text_box.pack(pady=20)

        self.record_button = tk.Button(self.root, text="Record", command=self.record_voice)
        self.record_button.pack(pady=20)

        self.recognizer = sr.Recognizer()
        self.engine = pyttsx3.init()

    def record_voice(self):
        with sr.Microphone() as source:
            self.text_box.insert(tk.END, "Listening...\n")
            self.root.update()
            audio = self.recognizer.listen(source)

        try:
            command = self.recognizer.recognize_google(audio)
            self.text_box.insert(tk.END, f"You said: {command}\n")
            self.process_command(command)
        except sr.UnknownValueError:
            self.text_box.insert(tk.END, "Sorry, I did not understand that.\n")
        except sr.RequestError:
            self.text_box.insert(tk.END, "Sorry, there was an error with the speech recognition service.\n")

    def process_command(self, command):
        response = ""
        if "hello" in command.lower():
            response = "Hello! How can I help you today?"
        elif "how are you" in command.lower():
            response = "I am just a program, so I don't have feelings, but thank you for asking!"
        elif "your name" in command.lower():
            response = "I am your voice assistant."
        else:
            response = "I am not sure how to respond to that."

        self.text_box.insert(tk.END, f"Assistant: {response}\n")
        self.speak(response)

    def speak(self, text):
        self.engine.say(text)
        self.engine.runAndWait()

if __name__ == "__main__":
    root = tk.Tk()
    assistant = VoiceAssistant(root)
    root.mainloop()
```
